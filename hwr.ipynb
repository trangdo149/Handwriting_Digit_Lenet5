```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, AveragePooling2D, Flatten, Dense
```

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print("Shape x_train: ", x_train.shape)
print("Shape y_train: ", y_train.shape)
print("Shape x_test: ", x_test.shape)
print("Shape y_test: ", y_test.shape)

x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
model = Sequential([
    Conv2D(6, kernel_size=(5,5), activation='tanh', input_shape=(28,28,1)),
    AveragePooling2D(pool_size=(2,2)),

    Conv2D(16, kernel_size=(5,5), activation='tanh'),
    AveragePooling2D(pool_size=(2,2)),
    Conv2D(120, kernel_size=(4,4), activation='tanh'),
    Flatten(),
    Dense(84, activation='tanh'),
    # Output Layer (10 class)
    Dense(10, activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train.reshape(-1,28,28,1),y_train,epochs=20)
```

```python
model.evaluate(x_test, y_test)
```

```python
import os
import cv2
model_path = "model_lenet5_v0.h5"

# Kiểm tra và tải model cũ
if os.path.exists(model_path):
    model = keras.models.load_model(model_path)
    print("\n Mô hình đã được tải lại thành công!")
else:
    print("\n Không tìm thấy model! Bạn cần huấn luyện lại mô hình trước.")
i = 0
while os.path.exists(f'img/{i}.png'):
    img = cv2.imread(f'img/{i}.png', cv2.IMREAD_GRAYSCALE)
    img = np.invert(np.array(img))
    img = img / 255.0
    img = img.reshape(1, 28, 28, 1)
    y_pred = model.predict(img)
    plt.imshow(img.squeeze(), cmap='gray')
    plt.show()
    print(f'Ket qua du doan la so: ', np.argmax(y_pred))
    i += 1
```
